## Method of estimating Binary Classification

- <h3>confusion matrix</h1>

<img src="https://cdn.prod.website-files.com/5d7b77b063a9066d83e1209c/63b413d2cdc133446aa23fc5_636b9182cfaef2115e028921_HERO_1_Confusion.png" style="width: 80%; height: 500px; background-color: none">

| Metric | Formula | Interpretation |
|---|---|---|
| Accuracy | $(TP + TN) \over (TP + TN + FP + FN)$ | Overall correct predictions |
| Precision | $TP \over (TP + FP)$ | Proportion of positive predictions that are correct |
| Recall | $TP \over (TP + FN)$ | Proportion of actual positive instances that were correctly predicted |
| F1-Score | $2 * (Precision * Recall) \over (Precision + Recall)$ | Balance between precision and recall |
| ROC Curve | Plot of TPR vs. FPR | Trade-off between true positive rate and false positive rate |
| AUC | Area under the ROC curve | Overall performance of the model |

- <h3>1. Accuracy</h3>

Accuracy is the proportion of all classifications that were **correct**, whether positive or negative. It is mathematically defined as: <br> $ (TP + TN) \over (TP + TN + FP + FN)$ <br> **rate of correct classification and total classicitaion**

A perfect model would have zero false positives and zero false negatives and therefore an accuracy of 1.0, or 100%.
However, when the dataset is imbalanced, or where one kind of mistake (FN or FP) is more costly than the other, which is the case in most real-world applications, it's better to optimize for one of the other metrics instead.

- ### 2. Recall
Recall = $(\frac{TP}{TP + FN})$
